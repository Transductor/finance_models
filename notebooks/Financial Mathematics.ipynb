```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def npv_from_tir(initial_investment, flow, term, irr):

    """
    Returns the NPV given project parameters

    Arguments:
    initial_investment - must be numeric float
    flow               - numpy array of cash flows of arbitrary size
    term               - compund interest term in number of days
    irr                - internal rate of return
    """

    n = len(flow)
    discount_vector = np.array([(1+irr*term/360)**j for j in range(1,n+1)])
    discounted_flow = np.divide(flow, discount_vector)
    return -initial_investment + np.sum(discounted_flow)
    
```

```python
#Example

horizon = 50
iv = 100
fl = np.array([float(i)*5 if i < 10 else 2 for i in range(horizon)])
t = 30.0
irr_vector = np.array([1+x/horizon for x in range(horizon)])
nvp_vector = np.array([npv_from_tir(iv, fl, t, irr) for irr in irr_vector])
```

```python
fig, ax = plt.subplots()

ax.plot(irr_vector,nvp_vector)
ax.set_xlabel("Internal rate of return")
ax.set_ylabel("NPV")
ax.set_ylim
ax.grid()
```
